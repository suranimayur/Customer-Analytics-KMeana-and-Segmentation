##### Libraries

```python
import numpy as np
import pandas as pd
import scipy 

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
import pickle
```

```python
df = pd.read_csv("segmentation data.csv", index_col =0)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

### Visualize Data

```python
plt.figure(figsize=(12,9))
plt.scatter(df.iloc[:,2],df.iloc[:,4])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Visualization of Raw Data")
```

#### Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

segmentation_std = scaler.fit_transform(df)
```

```python
## Hierarchial Clustering 
```

```python
hier_clust = linkage(segmentation_std,method ="ward")
```

```python
plt.figure(figsize=(12,9))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel("Observation")
plt.ylabel("Distance")

dendrogram(hier_clust, show_leaf_counts=False,
           truncate_mode="level",
           p = 5,
          no_labels=True
          )

plt.show()
```

## K- Mean Clustering

```python
wcss = []

for i in range(1,11):
    kmean = KMeans(n_clusters= i, init="k-means++", random_state=42)
    kmean.fit(segmentation_std)
    wcss.append(kmean.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss,marker='o',linestyle="--")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("K-Mean Clustering ")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4,init= "k-means++",random_state=42)
```

```python
kmeans.fit(segmentation_std)
```

### Results

```python
df_segm_kmeans = df.copy()
```

```python
df_segm_kmeans["Segment K-Means"] = kmeans.labels_
```

```python
df_segm_analysis = df_segm_kmeans.groupby(["Segment K-Means"]).mean()
```

```python
df_segm_analysis
```

```python
df_segm_analysis["N Obs"] = df_segm_kmeans[["Segment K-Means","Sex"]].groupby(["Segment K-Means"]).count()

df_segm_analysis["Pros Obs"] = df_segm_analysis["N Obs"]/df_segm_analysis["N Obs"].sum()
```

```python
df_segm_analysis
```

```python
df_segm_analysis.rename({0:"Well Off",
                            1:"Fewer Opportunity",
                            2:"Standard",
                            3:"Career Focussed"})
```

```python
df_segm_kmeans["Labels"]=df_segm_kmeans["Segment K-Means"].map({
    0:"Well-Off",
    1:"Fewer Opportunity",
    2: "Standard",
    3:"Carwwe Focused"
})
```

```python
df_segm_kmeans
```

```python
x_axis = df_segm_kmeans["Age"]
y_axis = df_segm_kmeans["Income"]
plt.figure(figsize=(10,8))
sns.scatterplot(x=x_axis,y=y_axis,hue=df_segm_kmeans["Labels"],palette=['g','r','c','m'])
plt.title('Segmentation K-Means')
plt.show()
```

## PCA

```python
pca = PCA()
```

```python
pca.fit(segmentation_std)
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(12,9))
plt.plot(range(1,8),pca.explained_variance_ratio_.cumsum(),marker='o',linestyle = '--')
plt.title("Explained Varience by Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Varience")
```

```python
pca = PCA(n_components=3)

pca.fit(segmentation_std)
```

### PCA Result

```python
pca.components_
```

```python
df_pca_comp = pd.DataFrame(data= pca.components_,
                           columns=df.columns.values,
                          index=["Component 1","Component 2","Component 3"])
```

```python
df_pca_comp
```

```python
sns.heatmap(df_pca_comp,vmax=1,vmin=-1,annot=True,cmap="RdBu"
           )

plt.yticks([0,1,2],
          ["Component 1","Component 2","Component 3"],
          rotation = -45,
          fontsize=15)
plt.show()
```

```python
pca.transform(segmentation_std)
```

```python
scores_pca = pca.transform(segmentation_std)
```

## K Mean Clustering 

```python
wcss = []

for i in range(1,11):
    kmean_pca = KMeans(n_clusters= i, init="k-means++", random_state=42)
    kmean_pca.fit(scores_pca)
    wcss.append(kmean_pca.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss,marker='o',linestyle="--")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("K-Mean with PCA Clustering ")
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters=4, init = "k-means++", 
                    random_state= 42)
```

```python
kmeans_pca.fit(scores_pca)
```

### K- Mean Clustering with PCA Results

```python
df_segm_pca_kmeans =pd.concat([df.reset_index(drop=True),pd.DataFrame(scores_pca)],axis=1)

df_segm_pca_kmeans.columns.values[-3:] = ["Component 1","Component 2","Component 3"]

df_segm_pca_kmeans["Segment K-means PCA"] = kmeans_pca.labels_
```

```python
df_segm_pca_kmeans
```

```python
df_segm_pca_kmeans_freq= df_segm_pca_kmeans.groupby(["Segment K-means PCA"]).mean()

df_segm_pca_kmeans_freq
```

```python
df_segm_pca_kmeans_freq["N Obs"] = df_segm_pca_kmeans[["Segment K-means PCA","Sex"]].groupby(["Segment K-means PCA"]).count()

df_segm_pca_kmeans_freq["Pros Obs"] = df_segm_pca_kmeans_freq["N Obs"]/df_segm_pca_kmeans_freq["N Obs"].sum()

df_segm_pca_kmeans_freq = df_segm_pca_kmeans_freq.rename({0:"standard",
                                                         1:"career focussed",
                                                         2: "fewer opportunities",
                                                         3:"well-off"})

df_segm_pca_kmeans_freq
```

```python
 df_segm_pca_kmeans["Legend"]= df_segm_pca_kmeans["Segment K-means PCA"].map({0:"standard",
                                                         1:"career focussed",
                                                         2: "fewer opportunities",
                                                         3:"well-off"})
```

```python
x_axis = df_segm_pca_kmeans["Component 2"]
y_axis = df_segm_pca_kmeans["Component 1"]
plt.figure(figsize=(10,9))
sns.scatterplot(x=x_axis,y=y_axis,hue= df_segm_pca_kmeans["Legend"],
               palette=["g","r","c","m"])
plt.title("Clusters by PCA Components")
plt.show()
```
